```python
for times in range(2):
    for row_index in range(row_amount):
        print(row_index, end='')
        for w_n, week_day in enumerate(list(operating_df[operating_df.columns[2:9]].loc[row_index])):
            if int(week_day) == 1:
                for t_n, time_perior in enumerate(list(operating_df[operating_df.columns[9:13]].loc[row_index])):
                    if int(time_perior) == 1:
                        #tmp_list = [0 for i in range(13)]
                        # gender and age /len=2
                        tmp_list = list(operating_df[operating_df.columns[0:2]].loc[row_index])
                        if tmp_list[0] == '男':
                            tmp_list[0] = 'M'
                        else:
                            tmp_list[0] = 'F'
                        if tmp_list[1] == '20-30歲':
                            tmp_list[1] = random.randint(20, 29)
                        elif tmp_list[1] == '未滿20歲':
                            tmp_list[1] = random.randint(12, 19)
                        elif tmp_list[1] == '30-40歲':
                            tmp_list[1] = random.randint(30, 39)
                        elif tmp_list[1] == '40-50歲':
                            tmp_list[1] = random.randint(40, 49)
                        elif tmp_list[1] == '50-60歲':
                            tmp_list[1] = random.randint(50, 59)   
                        else:
                            tmp_list[1] = random.randint(60, 80)


                        # weekday int 0~6 /len=1
                        tmp_list.append(w_n)

                        # time perior 5:00-10:00 10:00-15:00 15:00~20:00 other /len=1
                        rand_time_perior_list = [random.randint(5,10), random.randint(11,15), random.randint(15,20), [random.randint(0,4), random.randint(21,23)][random.randint(0,1)]]
                        tmp_list.append(rand_time_perior_list[t_n])

                        # ingredients /len=10
                        # 01234->13 , 5->5 , 6->11 , 7->9 , 8->10 , 9->8 , 10->4 , 11->6 , 12->12 , 13->7
                        tmp_ingredient_list = [0 for i in range(10)]
                        for in_n, ingredient_pick in enumerate(list(operating_df[operating_df.columns[13:27]].loc[row_index])):
                            if int(ingredient_pick) == 1:
                                if in_n == 0 or in_n == 1 or in_n == 2 or in_n == 3 or in_n == 4:
                                    tmp_ingredient_list[9] = 1
                                elif in_n == 5:
                                    tmp_ingredient_list[1] = 1
                                elif in_n == 6:
                                    tmp_ingredient_list[7] = 1
                                elif in_n == 7:
                                    tmp_ingredient_list[5] = 1
                                elif in_n == 8:
                                    tmp_ingredient_list[6] = 1
                                elif in_n == 9:
                                    tmp_ingredient_list[4] = 1
                                elif in_n == 10:
                                    tmp_ingredient_list[0] = 1
                                elif in_n == 11:
                                    tmp_ingredient_list[2] = 1
                                elif in_n == 12:
                                    tmp_ingredient_list[8] = 1
                                elif in_n == 13:
                                    tmp_ingredient_list[3] = 1
                        tmp_list += tmp_ingredient_list

                        ana_df = ana_df.append(pd.DataFrame([tmp_list], columns=ana_col), ignore_index=True)
ana_df
```
