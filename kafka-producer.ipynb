```python
'''

kafka producer

'''

if __name__ == "__main__":
    # 設定要連線到Kafka集群的相關設定, 並產生一個Kafka的Producer的實例
    producer = KafkaProducer(
        # Kafka集群在那裡?
#         bootstrap_servers=["35.220.196.201:9092"],
        bootstrap_servers=["10.170.0.3:9092"],
        # 指定msgKey的序列化器, 若Key為None, 無法序列化, 透過producer直接給值
        key_serializer=str.encode,
        # 指定msgValue的序列化器
        value_serializer=str.encode
    )

    # 指定想要發佈訊息的topic名稱
    #client = KafkaClient(hosts='34.80.44.128:9092')
    #topic_name = client.topics[2]
    topic_name = "U7173dc299af6ac3fe7697a75c2a280ad"
    topic_name = "Udf8f1c512878156d2bf8d0051d4f2429"

    topic_name_list = ['U7173dc299af6ac3fe7697a75c2a280ad', 
                       'Udf8f1c512878156d2bf8d0051d4f2429', 
                       'U1e91bf35eb5315bc17da624e46a65c51', 
                       'Uaefa84530bef60672e135c6cee89341d', 
                       'U9a1715c6ffdf0ef2ffe148726d336011', 
                       'Ua0ac6b70d3611ea9d3d4665941dbf3f7', 
                       'U30256c6d3c01f118f8c96f05e1f0f094', 
                       'U934211376c8da4cc677c3b801bd9f8db', 
                       'U26c1d062b5768c0ab8721698973d67b1',
                       'Uc132d7550826d16f3d86f60a24aaae20']
    # topic_name_list = ['U7173dc299af6ac3fe7697a75c2a280ad', 'Udf8f1c512878156d2bf8d0051d4f2429', 'U30256c6d3c01f118f8c96f05e1f0f094', 'U934211376c8da4cc677c3b801bd9f8db', 'U26c1d062b5768c0ab8721698973d67b1']
    # topic_name_list = ['U7173dc299af6ac3fe7697a75c2a280ad', 'Udf8f1c512878156d2bf8d0051d4f2429']
#     topic_name_list = ['U7173dc299af6ac3fe7697a75c2a280ad']

    def initiate_key_value(topic_name):
        for i in range(1, 11):
            print('init\t%s\t%s\t%s'%(topic_name, i, 0))
            producer.send(topic=topic_name, key=str(i), value="0")

    msg_counter = 0
    try:
        print("Start sending messages ...")
        
        #####
        while True:
            for p in range(index_amount):
                # 送給誰
                topic_name = topic_name_list[random.randint(0, len(topic_name_list) - 1)]                
                # 星期幾送資料
                if df_origin.loc[p][2] == time.localtime(time.time())[6]:
                    # 幾點送資料
                    if df_origin.loc[p][3] == time.localtime(time.time())[3]:
                        print('\n===============')
                        print('Weekday=%s'%(df_origin.loc[p][2]))
                        print('Time=%s'%(df_origin.loc[p][3]))
                        # 買了什麼
                        # 初始化00000
                        print('%s start transaction.'%(topic_name))
                        initiate_key_value(topic_name)
                        for item_num, bought in enumerate(df_origin.loc[p][4:14]):
                            if bought == 1:
                                producer.send(topic=topic_name, key=str(item_num + 1), value="1")
                                print('%s key=%s value=%s'%(topic_name, item_num + 1, 1))
                            time.sleep(1)
                        # 終止訊息
                        time.sleep(1)
                        producer.send(topic=topic_name, key="end", value="1")
                        print('%s finish transaction.'%(topic_name))
                        print('===============\n')
            
            # 每小時送一次資料
            print('pause 1 hr...')
            time.sleep(3500)

        print("Message sending completed!")
    except Exception as e:
        # 錯誤處理
        e_type, e_value, e_traceback = sys.exc_info()
        print("type ==> %s" % (e_type))
        print("value ==> %s" % (e_value))
        print("traceback ==> file name: %s" % (e_traceback.tb_frame.f_code.co_filename))
        print("traceback ==> line no: %s" % (e_traceback.tb_lineno))
        print("traceback ==> function name: %s" % (e_traceback.tb_frame.f_code.co_name))
    finally:
        # 關掉Producer實例的連線
        producer.close()
```
